```python
import pyperclip
from youtube_transcript_api import YouTubeTranscriptApi

# Function to get transcript
def get_transcript(video_id, languages=['en']):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages)
        return transcript
    except Exception as e:
        return str(e)

# Example video ID
video_id = 'DFsQNJG3qbk'

# Get and print transcript
transcript = get_transcript(video_id) #languages=['en-IN']

if isinstance(transcript, str):
    print(transcript)
else:
    my_text = ''
    for entry in transcript:
        my_text += entry['text'] + '\n'

    # Copy to clipboard
    pyperclip.copy(my_text)
    print("Transcript copied to clipboard.")
```
